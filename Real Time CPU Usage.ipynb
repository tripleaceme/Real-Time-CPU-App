```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from psutil import cpu_percent
%matplotlib notebook
#setting the time frame
frame_len = 200
# Our cpu percentage holding variable
y = []
#setting the fig size so that it wont cover much space
fig = plt.figure(figsize=(6,3))

#Declaring a function to loop through the cpu percentage and print it out
def animate(i):
    y.append(cpu_percent())
    
    if len(y) <= frame_len:
        plt.cla()
        #This is to let you know that you're below 200secs
        plt.plot(y,'r', label= 'Real-Time CPU Usage | Adegbite Ayoade')
    else:
        plt.cla()
        #This will change the color of the plot to blue so as to let you know that you've pass 200secs
        plt.plot(y[-frame_len:],'b', label= 'Real-Time CPU Usage | Adegbite Ayoade')
        
    plt.ylim(0,100)
    plt.xlabel("Time (s)")
    plt.ylabel("CPU Usage (%)")
    plt.legend(loc = "upper right")
    plt.tight_layout()
# the interval can be changed so that it prints after 1secs    
ani = FuncAnimation(plt.gcf(), animate , interval = 1000)
```
